# Converting the categorical variables into numerical data

```python
df.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
df
```

```python
#Now drop the columns which are irrelevent for survival predictions such PassengerId,Name,Ticket
```

```python
df.drop(columns={'PassengerId','Name','Ticket'},axis=1, inplace=True)
```

```python
df
```

#seprating features and target

```python
X = df.drop(columns='Survived', axis=1)
Y = df['Survived']
```

```python
print(X)
print(Y)
```

```python
#Splitting the data into training and testing
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
X_train.isnull().sum()
```

#Model training

```python
# we're using logistic regression model that uses binary classification for the prediction
```

```python
model = LogisticRegression()
```

```python
#training the model with trainign data
```

```python
model.fit(X_train,Y_train)
```

#Model Evaluation

```python
X_train_prediction=model.predict(X_train)
```

```python
from sklearn import metrics
```

```python
score = model.score(X_test,Y_test)
print(score)
```

```python
cm = metrics.confusion_matrix(Y_train, X_train_prediction)
print(cm)
```

```python
sns.heatmap(cm, annot = True, fmt = "d", square = True, cmap= "inferno")
plt.ylabel('Actual label')
plt.xlabel('predicted label')
title = ('Accuracy Score :',score)
plt.title(title, size = 10)
```
